**EXEMPLO 1**

O Hospital Santo Augusto de Anjo realiza mensalmente 3.000 transfusões de sangue em pacientes internados. Para que o hospital consiga manter seus estoques, são necessárias 60 doações de sangue por dia. A Tabela DistFreqVarQual.xlsx do excel apresenta o total de doadores para cada tipo sanguíneo em determinado dia. Construa a tabela de distribuição de 
frequências para o problema em questão. 

**Importa a biblioteca pandas**



```python
import pandas as pd
```

**Pega caminho dos dados**

```python
DadosExcel = '/content/drive/MyDrive/DistFreqVarQual.xlsx'
```

**Utiliza biblioteca pandas para receber dados da tabela excel**

```python
df = pd.read_excel(DadosExcel)
df
```

**Define o nome das colunas**

*F(i) = número de ocorrências de cada elemento*

```python
df.columns = ['Tipo Sanguíneo', 'F(i)']
df
```

**Faz a soma dos elementos F(i)**

```python
SomaFi = df['F(i)'].sum()
SomaFi
```

**Cria uma nova coluna em nosso DataFrame com o nome de F(ri)[%] e define seus valores**

*F(ri)[%] = Porcentagem relativa à frequência absoluta F(i)*

```python
df['F(ri)[%]'] = round(((df['F(i)']*100)/SomaFi),2)
df
```

**Cria uma nova coluna em nosso DataFrame com o nome de F(ai)e define seus valores**

*F(ai) = soma de todas as ocorrências até o elemento analisado*

*cumsum() = soma cumulativa*

```python
df['F(ai)'] = df['F(i)'].cumsum()
```

```python
df
```

**Cria uma nova coluna em nosso DataFrame com o nome de F(rai)[%] e define seus valores**

F(rai)[%] = Soma de todas as frequências relativas até o elemento analisado

```python
df['F(rai)[%]'] = df['F(ri)[%]'].cumsum()
df
```

**Cria uma linha depois da ultima linha com os dados passados em cada coluna**

```python
df.loc[len(df)] = ['Soma', df['F(i)'].sum(), df['F(ri)[%]'].sum(), '', '']
```
